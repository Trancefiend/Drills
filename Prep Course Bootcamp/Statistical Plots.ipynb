```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x = np.random.normal(10, 5, 1000) #Generating 1000 random values with a center of 10 and std dev of 5
plt.hist(x)

plt.show()
```

```python
#Generating 1000 random values with a center of 10 and std dev of 5
x = np.random.normal(10, 5, 1000) 
plt.hist(x, bins = 40, color = 'purple')
plt.title('Bins Demo')
plt.xlabel('Random Values')

plt.show()
```

```python
# More random data with different center.
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

plt.hist(x, bins = np.arange(-10, 40), alpha=.5, color = 'brown')
plt.hist(y, bins = np.arange(-10, 40), alpha=.5, color = 'blue')
plt.title('Manually Setting Bin Placement')
plt.xlabel('Random Values')

plt.show()
```

```python
# The same data as before, but normed using normed=True
plt.hist(x, normed=True, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(y, normed=True, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Normed histograms')
plt.xlabel('Random Values')

plt.show()
```
