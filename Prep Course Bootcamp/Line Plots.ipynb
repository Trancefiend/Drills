```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

```

```python
plt.plot([1, 2, 3])
plt.show()
```

```python
np.random.seed(1)

df = pd.DataFrame()

df['random'] = np.random.rand(100)
df['random_sqr'] = df['random'] ** 2
df['random_shift'] = df['random'] + 2

df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['random'], color = 'green')
plt.title('Random Series')

plt.show()
```

```python
plt.plot(df['random'], color = 'green')
plt.plot(df['random_shift'], color = 'red')
plt.ylabel('Values')
plt.xlabel('Count')
plt.ylim([-0.4, 3.5])
plt.show()
```

```python
plt.scatter(x=df['random'], y=df['random_sqr'], color = 'purple')
plt.show()
```
