```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
bloodpressure = pd.read_csv('bloodpressure.csv', low_memory = False)
bloodpressure.dropna(subset=['BPXSY1'], inplace=True)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(bloodpressure['BPXSY1'], bins = 40, color = 'blue')
plt.title('Systolic Blood Pressure')
plt.xlabel('mmHg')
plt.ylabel('Counts')

plt.subplot(1, 2, 2)
plt.hist(bloodpressure['BPXDI1'], bins = 40, color = 'red')
plt.title('Diastolic Blood Pressure')
plt.xlabel('mmHg')
plt.ylabel('Counts')

plt.show()
```

In the above graphs you see the differences in distribution between systolic and diastolic blood pressures in a population. As you can see, the systolic pressure has a long tail to the right, whereas the diastolic blood pressures follow a more normal distribution. This matches with what modern medicine has learned about blood pressures.

```python
bloodpressure.dropna(subset=['BPXDI1'], inplace=True)
```

```python
plt.hist(bloodpressure['BPXDI1'], bins = 40, color = 'red')
plt.title('Diastolic Blood Pressure')
plt.xlabel('mmHg')
plt.ylabel('Counts')

plt.show()
```

```python
bloodpressure.dropna(subset=['BPXPLS'], inplace=True)
```

```python
plt.scatter(x=bloodpressure['BPXSY1'], y=bloodpressure['BPXPLS'], color = 'green')
plt.title('Blood Pressue vs Pulse Rate')
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('BPM')

plt.show()
```

In the above scatterplot, we are comparing systolic blood pressure levels to BPMs. As you can see there is no positive correlation between an in increase in blood pressure and an increase in BPM, as high blood pressures do not result in higher or lower BPMs. The same can be said for the relationship between BPM and blood pressure.

```python
no_food = bloodpressure.loc[lambda df: bloodpressure['BPQ150A'] == 1]
yes_food = bloodpressure.loc[lambda df: bloodpressure['BPQ150A'] == 2]
```

```python
data = [no_food['BPXSY1'], yes_food['BPXSY1']]
plt.boxplot(data)
plt.title('Blood Pressure and Food')
plt.ylabel('mmHg')
labels = ['', 'No Food', 'Food 30 Min']
plt.xticks(np.arange(3), labels)


plt.show()
```

Finally, in the above boxplots, we are looking at the difference in distribution between blood pressures taken without the patient eating before, and the patient eating less than 30 minutes prior to the measurement. As you can see, there is more variance in the blood pressures of taken from people who had recently ate. This confirms what we know about the relationship between eating and blood pressure levels.
